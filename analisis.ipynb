```python
# Load the uploaded Excel file to analyze the data and create necessary plots.
import pandas as pd
import matplotlib.pyplot as plt
import openpyxl as pxl
# Load the Excel data into a DataFrame
data_path = '\datos01.xls'
data_df = pd.read_excel('dataxlsx.xlsx')

# Display the first few rows to understand the structure of the data
data_df.head()
```

```python
# Filtrar los datos por cada experimento y crear grficos
experimentos = data_df['nExperimento'].unique()

for experimento in experimentos:
    df_experimento = data_df[data_df['nExperimento'] == experimento]
    plt.scatter(df_experimento['i'], df_experimento['tiempoABB'], label='tiempoABB', color='blue')
    plt.scatter(df_experimento['i'], df_experimento['tiempoSplay'], label='tiempoSplay', color='red')
    plt.xlabel('i')
    plt.ylabel('Tiempo')
    plt.title(f'Experimento {experimento}')
    plt.legend()
    plt.show()
```

```python
# Calcular el costo promedio para cada experimento y crear grficos
for experimento in experimentos:
    df_experimento = data_df[data_df['nExperimento'] == experimento]
    plt.plot(df_experimento['N'], df_experimento['cPromABB'], label='cPromABB', color='blue')
    plt.plot(df_experimento['N'], df_experimento['cPromSplay'], label='cPromSplay', color='red')
    plt.xlabel('N')
    plt.ylabel('Costo Promedio')
    plt.title(f'Costo Promedio por Experimento {experimento}')
    plt.legend()
    plt.show()
```
